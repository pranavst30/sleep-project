Random forest

```
def classification_table(y_true, y_pred, class_names):
    report = classification_report(
        y_true, y_pred,
        target_names=class_names,
        output_dict=True
    )

    rows = []
    index = []
    total_support = 0

    for cls in class_names:
        r = report[cls]
        rows.append([
            round(r["precision"] * 100, 2),
            round(r["recall"] * 100, 2),
            round(r["f1-score"] * 100, 2),
            int(r["support"])
        ])
        index.append(cls)
        total_support += int(r["support"])

    # accuracy
    rows.append(["-", "-", round(report["accuracy"] * 100, 2), total_support])
    index.append("accuracy")

    # macro avg
    macro = report["macro avg"]
    rows.append([
        round(macro["precision"] * 100, 2),
        round(macro["recall"] * 100, 2),
        round(macro["f1-score"] * 100, 2),
        total_support
    ])
    index.append("macro avg")

    # weighted avg
    weighted = report["weighted avg"]
    rows.append([
        round(weighted["precision"] * 100, 2),
        round(weighted["recall"] * 100, 2),
        round(weighted["f1-score"] * 100, 2),
        total_support
    ])
    index.append("weighted avg")

    return pd.DataFrame(
        rows,
        index=index,
        columns=["precision", "recall", "f1-score", "support"]
    )
```

```
class_names = ["Healthy", "Insomnia", "Sleep Apnea"]
```

```
rf = RandomForestClassifier(random_state=42)
rf.fit(X_train, y_train)
```

```
y_train_pred = rf.predict(X_train)

print("\n=== Overall Training Performance Comparison (Without Optimization) ===")
df_train = classification_table(
    y_train, y_train_pred, class_names
)
print(df_train)
```

```
y_test_pred = rf.predict(X_test)

print("\n=== Overall Testing Performance Comparison (Without Optimization) ===")
df_test_no_opt = classification_table(
    y_test, y_test_pred, class_names
)
print(df_test_no_opt)
```

```
y_cv_pred = cross_val_predict(
    rf,
    X_train,
    y_train,
    cv=10
)

print("\n=== Overall Cross-Validation (CV) Training Performance Comparison ===")
df_cv = classification_table(
    y_train, y_cv_pred, class_names
)
print(df_cv)
```

```
def objective(trial):
    model = RandomForestClassifier(
        n_estimators=trial.suggest_int("n_estimators", 100, 400),
        max_depth=trial.suggest_int("max_depth", 2, 20),
        max_features=trial.suggest_categorical(
            "max_features", ["sqrt", "log2", None]
        ),
        random_state=42
    )
    return cross_val_score(
        model, X_train, y_train,
        cv=10, scoring="accuracy"
    ).mean()

optuna.logging.set_verbosity(optuna.logging.WARNING)

study = optuna.create_study(direction="maximize")
study.optimize(objective, n_trials=30)
```

```
best_rf = RandomForestClassifier(
    **study.best_params,
    random_state=42
)
best_rf.fit(X_train, y_train)
```

```
y_test_pred_opt = best_rf.predict(X_test)

print("\n=== Overall Testing Performance Comparison (After Optimization) ===")
df_test_opt = classification_table(
    y_test, y_test_pred_opt, class_names
)
print(df_test_opt)
```

```
import pandas as pd
import optuna

from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.discriminant_analysis import QuadraticDiscriminantAnalysis
from sklearn.model_selection import cross_val_score, cross_val_predict
from sklearn.metrics import accuracy_score, classification_report
```

```
def classification_table(y_true, y_pred, class_names):
    accuracy = accuracy_score(y_true, y_pred) * 100
    total_samples = len(y_true)

    report = classification_report(
        y_true, y_pred,
        target_names=class_names,
        output_dict=True
    )

    df = pd.DataFrame(report).transpose()
    df = df.drop(index="accuracy", errors="ignore")

    for col in ["precision", "recall", "f1-score"]:
        df[col] = (df[col] * 100).round(2)

    df["support"] = df["support"].astype(int)

    df.loc["accuracy"] = ["-", "-", f"{accuracy:.2f}", total_samples]

    return df[["precision", "recall", "f1-score", "support"]]
```

```
class_names = ['Healthy', 'Sleep Apnea', 'Insomnia']

qda_base = Pipeline([
    ("scaler", StandardScaler()),
    ("qda", QuadraticDiscriminantAnalysis())
])

qda_base.fit(X_train, y_train)
```

```
y_train_pred = qda_base.predict(X_train)

print("\n=== Overall Training Performance Comparison (QDA – No Optimization) ===")
df_train = classification_table(
    y_train, y_train_pred, class_names
)
print(df_train)
```

```
y_test_pred = qda_base.predict(X_test)

print("\n=== Overall Testing Performance Comparison (QDA – No Optimization) ===")
df_test_no_opt = classification_table(
    y_test, y_test_pred, class_names
)
print(df_test_no_opt)
```

```
y_cv_pred = cross_val_predict(
    qda_base,
    X_train,
    y_train,
    cv=10
)

print("\n=== Overall Cross-Validation (CV) Training Performance Comparison (QDA) ===")
df_cv = classification_table(
    y_train, y_cv_pred, class_names
)
print(df_cv)
```

```
def objective(trial):
    model = Pipeline([
        ("scaler", StandardScaler()),
        ("qda", QuadraticDiscriminantAnalysis(
            reg_param=trial.suggest_float("reg_param", 0.0, 1.0),
            tol=trial.suggest_float("tol", 1e-5, 1e-2, log=True),
            store_covariance=trial.suggest_categorical(
                "store_covariance", [True, False]
            )
        ))
    ])

    return cross_val_score(
        model, X_train, y_train,
        cv=10, scoring="accuracy"
    ).mean()

optuna.logging.set_verbosity(optuna.logging.WARNING)

study = optuna.create_study(direction="maximize")
study.optimize(objective, n_trials=50)

print("\nBest Hyperparameters from Bayesian Optimization:")
print(study.best_params)
```

```
qda_tuned = Pipeline([
    ("scaler", StandardScaler()),
    ("qda", QuadraticDiscriminantAnalysis(**study.best_params))
])

qda_tuned.fit(X_train, y_train)
```

```
import pandas as pd
import lightgbm as lgb
import optuna

from sklearn.model_selection import cross_val_score, cross_val_predict
from sklearn.metrics import accuracy_score, classification_report
```

```
class_names = ['Healthy', 'Sleep Apnea', 'Insomnia']
```

```
def classification_table(y_true, y_pred, class_names):
    accuracy = accuracy_score(y_true, y_pred) * 100
    total_samples = len(y_true)

    report = classification_report(
        y_true, y_pred,
        target_names=class_names,
        output_dict=True
    )

    df = pd.DataFrame(report).transpose()
    df = df.drop(index='accuracy', errors='ignore')

    for col in ['precision', 'recall', 'f1-score']:
        df[col] = (df[col] * 100).round(2)

    df['support'] = df['support'].astype(int)
    df.loc['Accuracy'] = ['-', '-', f"{accuracy:.2f} %", total_samples]

    return df[['precision', 'recall', 'f1-score', 'support']]
```

```
categorical_cols = ['Gender', 'Occupation', 'BMI Category', 'Blood Pressure']

for col in categorical_cols:
    X_train[col] = X_train[col].astype('category')
    X_test[col] = X_test[col].astype('category')
```

```
lgb_base = lgb.LGBMClassifier(
    random_state=42,
    verbosity=-1
)

lgb_base.fit(
    X_train,
    y_train,
    categorical_feature=categorical_cols
)
```

```
y_train_pred = lgb_base.predict(X_train)

print("\n=== Overall Training Performance Comparison (LightGBM – No Optimization) ===")
df_train = classification_table(
    y_train, y_train_pred, class_names
)
print(df_train)
```

```
import pandas as pd
import numpy as np
import optuna
import warnings
warnings.filterwarnings("ignore")

from sklearn.model_selection import StratifiedKFold, cross_val_score, cross_val_predict
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report
```

```
categorical_cols = ['Gender', 'Occupation', 'BMI Category', 'Blood Pressure']
numeric_cols = [col for col in X_train.columns if col not in categorical_cols]

preprocessor = ColumnTransformer([
    ('onehot', OneHotEncoder(drop='first', handle_unknown='ignore'), categorical_cols),
    ('scale', StandardScaler(), numeric_cols)
])
```

```
class_names = ['Healthy', 'Sleep Apnea', 'Insomnia']
```

```
def classification_table(y_true, y_pred, class_names):
    acc = accuracy_score(y_true, y_pred) * 100
    total = len(y_true)

    report = classification_report(
        y_true, y_pred,
        target_names=class_names,
        output_dict=True
    )

    df = pd.DataFrame(report).transpose()
    df = df.drop(index='accuracy', errors='ignore')

    for col in ['precision', 'recall', 'f1-score']:
        df[col] = (df[col] * 100).round(2)

    df['support'] = df['support'].astype(int)
    df.loc['Accuracy'] = ['-', '-', f"{acc:.2f} %", total]

    return df[['precision', 'recall', 'f1-score', 'support']]
```

```
svm_base = Pipeline([
    ('preprocess', preprocessor),
    ('svm', SVC(
        class_weight='balanced',
        random_state=42
    ))
])

svm_base.fit(X_train, y_train)
```

```
y_train_pred = svm_base.predict(X_train)

print("\n=== Overall Training Performance Comparison (SVM – No Optimization) ===")
print(classification_table(y_train, y_train_pred, class_names))
```

```
y_test_pred = svm_base.predict(X_test)

print("\n=== Overall Testing Performance Comparison (SVM – No Optimization) ===")
print(classification_table(y_test, y_test_pred, class_names))
```

```
cv = StratifiedKFold(n_splits=10, shuffle=True, random_state=42)

y_cv_pred = cross_val_predict(
    svm_base,
    X_train,
    y_train,
    cv=cv,
    n_jobs=-1
)

print("\n=== Overall Cross-Validation (CV) Training Performance Comparison (SVM) ===")
print(classification_table(y_train, y_cv_pred, class_names))
```

```
# --- Final SVM Decision Boundary Visualization (Cleaned) ---

import matplotlib.pyplot as plt
from sklearn.svm import SVC
import numpy as np
import pandas as pd

# Select only needed columns
data = df[['Stress Level', 'Physical Activity Level', 'Sleep Disorder']].copy()

# Clean target column: replace blanks or unknowns with NaN, then drop them
data['Sleep Disorder'] = data['Sleep Disorder'].replace(['', ' ', 'Unknown', 'NaN', np.nan], np.nan)
data = data.dropna(subset=['Sleep Disorder', 'Stress Level', 'Physical Activity Level'])

# Encode target labels safely
label_map = {'None': 0, 'Insomnia': 1, 'Sleep Apnea': 2}
data = data[data['Sleep Disorder'].isin(label_map.keys())]  # keep only valid rows
data['Sleep Disorder'] = data['Sleep Disorder'].map(label_map)

# Extract features and labels
X = data[['Stress Level', 'Physical Activity Level']].values
y = data['Sleep Disorder'].values

# Train SVM
model = SVC(kernel='rbf', gamma='auto')
model.fit(X, y)

# Create meshgrid for visualization
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.linspace(x_min, x_max, 500),
                     np.linspace(y_min, y_max, 500))

# Predict for each grid point
Z = model.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# Plot decision boundary
plt.figure(figsize=(8,6))
plt.contourf(xx, yy, Z, alpha=0.3, cmap='coolwarm')
plt.scatter(X[:,0], X[:,1], c=y, s=30, edgecolors='k', cmap='coolwarm')
plt.xlabel('Stress Level')
plt.ylabel('Physical Activity Level')
plt.title('Diagram 3: 2D Decision Boundary Plot (SVM)')
plt.show()
```
